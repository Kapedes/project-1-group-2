```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt

# File to Load (Remember to Change These)
crash_data = Path("Crash_Data.csv")

# Import the API key
from config import api_key
```

```python
crash_data_df = pd.read_csv(crash_data)

crash_data_df
```

```python
#Choosing data fields
cleaned_crash_data_df = crash_data_df[["State", "Month", "Year", "Time", "Dayweek", "Day of week", "Time of day", "Road User", "Gender", "Age Group", "Christmas Period", "Easter Period"]]
cleaned_crash_data_df
```

```python
#Dropping Na's

cleaned_crash_data_1_df = cleaned_crash_data_df.dropna(how="any") 

cleaned_crash_data_1_df
```

```python
#Replacing state names

cleaned_crash_data_1_df['State'] = cleaned_crash_data_1_df['State'].replace({"Vic": "VIC", "Qld": "QLD", "Tas": "TAS"})

cleaned_crash_data_1_df
```

```python
#dataset 2 used
cleaned_crash_data_2_df = crash_data_df[["State", "Month", "Year", "Time", "Dayweek", "Day of week", "Time of day", "Road User", "Gender", "Age Group", "National Road Type", "Christmas Period", "Easter Period"]]

cleaned_crash_data_2_df
```

```python
# Dataset 2 with dropped Na's 

cleaned_crash_data_2_df = cleaned_crash_data_df.dropna(how="any") 

cleaned_crash_data_2_df
```

```python
#Dataset 2 changed names
cleaned_crash_data_2_df['State'] = cleaned_crash_data_2_df['State'].replace({"Vic": "VIC", "Qld": "QLD", "Tas": "TAS"})

cleaned_crash_data_2_df
```

```python
#FATAL CRASHES BY AGE GROUPS BY DRIVER NOT GROSSED BY AGE GROUP

data_driver_age = cleaned_crash_data_2_df[cleaned_crash_data_2_df['Road User'] == 'Driver']

Age_of_driver = data_driver_age.groupby('Age Group').size()

Age_of_driver.plot(kind = "pie" , autopct="%1.1f%%")
plt.ylabel('Age of Driver')
plt.title('Fatalities by Drivers by Age Group (Not normalised)')
plt.show()
```

```python
#DIVIDING THE VALUES TO WHICH MOST FATALITIES OCCUR FOR DRIVERS BY AGE GROUP 
# Performing steps to divide the groups by number of ages in age group
Revised_driver_ages = pd.DataFrame(Age_of_driver)
#Renaming column fatalities
Revised_driver_ages = Revised_driver_ages.rename(columns={ 0: 'Fatalities'})
Rev_Driver_age = pd.DataFrame(Revised_driver_ages)


# Define division values for each Age Group
division_values = {
    '0_to_16': 3,
    '17_to_25': 9,
    '26_to_39': 14,
    '40_to_64': 15,
    '65_to_74': 9,
    '75_or_older': 18
}

# Divide Fatalities by the specified values for each Age Group
Rev_Driver_age['Fatalities'] = Rev_Driver_age.apply(lambda row: row['Fatalities'] / division_values[row.name], axis=1)

# Display the modified DataFrame
print(Rev_Driver_age)

```

```python
#Creating a list for to use ina  subsequent procedure
List_age_group = cleaned_crash_data_1_df['Age Group'].unique()
List_age_group.sort()
List_age_group
```

```python

#Creating dataframe
Driver_Age_Fatalities = pd.DataFrame(Rev_Driver_age,index=List_age_group)

# Create a pie chart
plt.figure(figsize=(4, 4))
plt.pie(Driver_Age_Fatalities['Fatalities'], labels=Driver_Age_Fatalities.index, autopct='%1.1f%%', startangle=0)
plt.title('Fatalities by Drivers by Age Group (Normalised)')
plt.show()
```

```python
# Getting an API to get population

import requests
# List of Australian capital cities
australian_capital_cities = [
    'Canberra',
    'Sydney',
    'Melbourne',
    'Brisbane',
    'Perth',
    'Adelaide',
    'Hobart',
    'Darwin'
]

base_url = 'https://api.api-ninjas.com/v1/city?name={}'
for city in australian_capital_cities:
    api_url = base_url.format(city)
    response = requests.get(api_url, headers={'X-Api-Key': api_key})
    if response.status_code == requests.codes.ok:
        data = response.json()  # Assuming the response is in JSON format
        if data and isinstance(data, list) and len(data) > 0:
            city_data = data[0]  # Get the first item in the response
            population = city_data.get('population', 'Population data not available')
            print(f"{city}: Population - {population}")
        else:
            print(f"No data available for {city}")
    else:
        print(f"Error fetching data for {city}: {response.status_code} - {response.text}")

```

```python
### Marianne's section 

### What time of day has the most fatal crashes in Victoria?

# Convert time column to "datetime" format for easy analysis
cleaned_crash_data_2_df['Time'] = pd.to_datetime(cleaned_crash_data_2_df['Time'], format='%H:%M')

# Extract the hour from the 'Time' column and create a new 'Hour' column
cleaned_crash_data_2_df['Hour'] = cleaned_crash_data_2_df['Time'].dt.hour

# Filter the data for rows where the State is 'VIC'
data_vic = cleaned_crash_data_2_df[cleaned_crash_data_2_df['State'] == 'VIC']

# Group by the 'Hour' column, count the number of crashes in each hour, and sort the result
hourly_crash_counts = data_vic.groupby('Hour').size().sort_values(ascending=False)

# Convert the series to a DataFrame for a nicer display
hourly_crash_counts_df = hourly_crash_counts.reset_index(name='Crash_Counts')
hourly_crash_counts_df.columns = ['Hour', 'Crash_Counts']

# Display the DataFrame
print(hourly_crash_counts_df)
```

```python
# plot this as a bar graph

# Set the size
plt.figure(figsize=(10, 6))

# Create a bar chart
# 'Hour' is on the x-axis and 'Crash_Counts' is on the y-axis
plt.bar(hourly_crash_counts_df['Hour'], hourly_crash_counts_df['Crash_Counts'], color='skyblue')

# Add labels and title with updated information
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Fatal Crashes')
plt.title('Number of Fatal Crashes by Hour of the Day in Victoria from 1989 - 2021')

# Set x-ticks and x-tick labels to improve readability
plt.xticks(hourly_crash_counts_df['Hour'], [f'{hour}:00' for hour in hourly_crash_counts_df['Hour']])

# Optionally, add a grid
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Rotate the x-ticks for better visibility
plt.xticks(rotation=45)

# Show the plot
plt.show()
```
