```python
# Dependencies and Setup
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# File to Load (Remember to Change These)
crash_data = "C:/Users/maria/OneDrive/Desktop/Data Analyst Course/Week 7 - Project 1/project-1-group-2/Crash Data/Crash_Data.csv"
```

```python
crash_data_df = pd.read_csv(crash_data)

crash_data_df
```

```python
cleaned_crash_data_df = crash_data_df[["State", "Month", "Year", "Time", "Dayweek", "Day of week", "Time of day", "Road User", "Gender", "Age Group", "Christmas Period", "Easter Period"]]

cleaned_crash_data_df
```

```python
cleaned_crash_data_1_df = cleaned_crash_data_df.dropna(how="any") 

cleaned_crash_data_1_df
```

```python
# The following is the first dataset to use

cleaned_crash_data_1_df['State'] = cleaned_crash_data_1_df['State'].replace({"Vic": "VIC", "Qld": "QLD", "Tas": "TAS"})

cleaned_crash_data_1_df
```

```python
cleaned_crash_data_2_df = crash_data_df[["State", "Month", "Year", "Time", "Dayweek", "Day of week", "Time of day", "Road User", "Gender", "Age Group", "National Road Type", "Christmas Period", "Easter Period"]]

cleaned_crash_data_2_df
```

```python
cleaned_crash_data_2_df = cleaned_crash_data_df.dropna(how="any") 

cleaned_crash_data_2_df
```

```python
# The following is the second dataset to use

cleaned_crash_data_2_df['State'] = cleaned_crash_data_2_df['State'].replace({"Vic": "VIC", "Qld": "QLD", "Tas": "TAS"})

cleaned_crash_data_2_df
```

```python
### Marianne's section 

### What time of day has the most fatal crashes in Victoria?

# Convert time column to "datetime" format for easy analysis
cleaned_crash_data_2_df['Time'] = pd.to_datetime(cleaned_crash_data_2_df['Time'], format='%H:%M')

# Extract the hour from the 'Time' column and create a new 'Hour' column
cleaned_crash_data_2_df['Hour'] = cleaned_crash_data_2_df['Time'].dt.hour

# Filter the data for rows where the State is 'VIC'
data_vic = cleaned_crash_data_2_df[cleaned_crash_data_2_df['State'] == 'VIC']

# Group by the 'Hour' column, count the number of crashes in each hour, and sort the result
hourly_crash_counts = data_vic.groupby('Hour').size().sort_values(ascending=False)

# Convert the series to a DataFrame for a nicer display
hourly_crash_counts_df = hourly_crash_counts.reset_index(name='Crash_Counts')
hourly_crash_counts_df.columns = ['Hour', 'Crash_Counts']

# Display the DataFrame
print(hourly_crash_counts_df)
```

```python
# plot this as a bar graph

# Set the size
plt.figure(figsize=(10, 6))

# Create a bar chart
# 'Hour' is on the x-axis and 'Crash_Counts' is on the y-axis
plt.bar(hourly_crash_counts_df['Hour'], hourly_crash_counts_df['Crash_Counts'], color='skyblue')

# Add labels and title with updated information
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Fatal Crashes')
plt.title('Number of Fatal Crashes by Hour of the Day in Victoria from 1989 - 2021')

# Set x-ticks and x-tick labels to improve readability
plt.xticks(hourly_crash_counts_df['Hour'], [f'{hour}:00' for hour in hourly_crash_counts_df['Hour']])

# Optionally, add a grid
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Rotate the x-ticks for better visibility
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

```python
# Count the occurrences of each gender in the 'Gender' column
gender_counts = data_vic['Gender'].value_counts()

# Display the counts
print(gender_counts)
```

```python
# Filter data for rows where 'Road User' is 'Driver'
driver_data = data_vic[data_vic['Road User'] == 'Driver']

# Count the amount of each gender in the 'Gender' column 
gender_counts_driver = driver_data['Gender'].value_counts()

# Create a bar chart
gender_counts_driver.plot(kind='bar', color=['blue', 'red'])

# Add labels and title
plt.xlabel('Gender')
plt.ylabel('Number of Crashes')
plt.title('Number of Crashes by Gender by Driver')

# Show the plot
plt.show()
```
